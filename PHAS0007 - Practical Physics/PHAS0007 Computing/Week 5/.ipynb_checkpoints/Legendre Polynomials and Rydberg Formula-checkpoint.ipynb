# Legendre Polynomials and Rydberg Formula
In this notebook, I will be plotting the values for the First, third and fifth Legendre polynomials between x values of -1 and 1. Further To this, I will also be calculating values for the first 4 wavelengths of light from the Rydberg Formula of each series.


## Legendre Polynomials
In this first section I will be plotting the first, third and fifth Legendre Polynomials on a graph range from -1 to 1. In the cell below I will first import any libraries that are required:

```python
#-----------------------------------------------------------------------
# Date: 01/11/22 Code Written by: Konstantinos Doran
#-----------------------------------------------------------------------
#import necessary libraries
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Now that we have our libraries, I will create 3 seperate lines plotted on the same graph:

```python
#create array with 500 values from -1 to 1 so smooth graph
xline = np.linspace(-1,1,500)
#define polynomials
first_y_line = xline
thirdyline = 0.5*(5*(xline**3)-3*xline)
fifthyline = 0.125*((63*(xline**5))-(70*(xline**3))+(15*xline))
#plot the graph
plt.figure(dpi = 150)
#add a grid
plt.grid(True)
#add labels and titles
plt.title("Legendre Polynomials")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
#plot 3 lines
plt.plot(xline,firstyline, label = "$P_1(x)$")
plt.plot(xline,thirdyline, label = "$P_3(x)$")
plt.plot(xline, fifthyline, label = "$P_5(x)$")
plt.legend(loc="best")
```

## Rydberg Formula

### First Value of Balmer Series:
First, I will be using the Rydberg Formula:
$$\frac{1}{\lambda} = R_H \left( \frac{1}{n_1^2} - \frac{1}{n_2^2} \right)$$
Which will give us the wavelength of the light emitted when an electron transitions from energy level $n_1$ to $n_2$ (with $n_2 > n_1$). For the Balmer series, $n_1 = 2$ and the first wavelength will be when $n_2 = 3$. Thus, we can substitute in these values and rearrange for the wavelength:

```python
# define constants
Rydberg = (1.09677583*10**7) #m^-1
n1 = 2 #unitless
n2 = 3 #unitless
#calculate wavelength
recipwave = Rydberg*(((1/n1)**2)-((1/n2)**2)) #m^-1
wavelength = 1/recipwave #m
nanowave = wavelength * 10**9 #nm
#output wavelength
#print(wavelength)
#print(nanowave)
print(f"The wavelength of the first line in the Balmer Series is: {nanowave:.0f}nm")
```
