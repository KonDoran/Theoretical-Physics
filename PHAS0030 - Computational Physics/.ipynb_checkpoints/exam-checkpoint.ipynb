```python
def charge_step(position,velocity,Bfield,dt):
    """Takes in positions and velocities of moving charged particle in a magnetic
    field and calculates new positions and velocities after one time step.
    Inputs:
    position   position vector (1D array) of charged particle
    velocity   velcotiy vector (1D array) of charged particle
    Bfield     magnetic field vector (1D array)
    dt         time step
    Outputs:
    newpos     new position vector
    newvel     new velocity vector
    """
    q = 1
    m = 1
    force = (dt/2)*(q*Bfield/m)
    G = 2*force/(1+(np.linalg.norm(force))**2)
    half_pos = position + dt/2 *velocity
    half_vel = velocity + np.cross(velocity, force)
    newvel = velocity + np.cross(half_vel, G)
    newpos = half_pos + (dt/2)*newvel
    return newpos, newvel
```

## b)

```python
Nsteps = 20000
positions = np.zeros((Nsteps+1,3))
velocities = np.zeros((Nsteps+1,3))
positions[0] = np.array([1.0,0.0,0.0])
velocities[0] = np.array([0.1,0.0,0.0])
dt = 0.01
t_arr = np.arange(0,200+dt,dt)
nu = 0.01
Bfield = np.array([0,0,np.cos(2*np.pi*nu*t_arr[0])])
for i in range(Nsteps):
    positions[i+1], velocities[i+1] = charge_step(positions[i],velocities[i],Bfield,dt)
    Bfield = np.array([0,0,np.cos(2*np.pi*nu*t_arr[i+1])])
```

```python
def psinorm(psi, dx):
    """Normalises a real wave function defined on a grid with grid spacing dx
    Inputs:
    psi     wave function as an array
    dx      grid spacing 
    Outputs:
    psinorm normalised wavefunction
    """
    integral = 0.0
    N = len(psi)
    for i in range(N):
        integral+= (psi[i])**2
    integral = integral*dx
    psinorm = psi/(np.sqrt(integral))
    return psinorm
        
    
```

```python
def secondforwdiffderivative(psi,dx):
    """Takes in function and calculates second derivative when defined on a grid using centred difference formula
    Inputs:
    psi     wave function as an array
    dx      grid spacing 
    Outputs:
    d2psi   second derivative of wave function
    """
    psi_p1 = np.roll(psi,1)
    psi_m1 = np.roll(psi,-1)
    d2psi = (psi_p1 -2*psi + psi_m1)/(dx*dx)
    return d2psi
```

For this function I have assumed that the function is recurring for the interval I have specified or that the function tends to zero at both boundaries.

```python
def energy(psi,dx,potential):
    """Calculates the energy of the wavefunction
    Inputs:
    psi        wave function as an array
    dx         grid spacing 
    potential  potential energy as an array
    Output:
    energy     total energy of wave function
    """
    KE = -0.5*secondforwdiffderivative(psi,dx)
    integral = 0.0
    N = len(psi)
    for i in range(N):
        integral += psi[i]*(KE[i]+potential[i])*psi[i]
    energy = integral*dx
    return energy
        
```
